```python
#Let's show the number of categorical data as a bar chart
```

```python
#Gender Distribution
```

```python
sns.countplot(x='sex', data=df)
plt.title('Gender Distribution')  #Add a title to the chart
plt.xlabel("Sex") #x axis name
plt.ylabel("Count")#y axis name
```

```python
#smoker
```

```python
sns.countplot(x='smoker',data=df)
```

```python
#Visualizing data in column "sex" relative to column "smoker"
```

```python
sns.countplot(x='sex', hue='smoker', data=df)
plt.title('Gender and Smoking Status')
```

```python
#Comparing numerical data with categorical data  ---------barplot()
```

```python
sns.barplot(x='sex', y='total_bill', data=df)
plt.title('Gender and Total Bill')
```

```python
#When we examine the chart, we can see the average bill amount paid according to gender distribution.
```

```python
#let's digitize this
```

```python
#Average of total invoice amounts of male customers
```

```python
df[df["sex"]=="Male"]["total_bill"].mean()
```

```python
#Average of total invoice amounts of customers
```

```python
df[df["sex"]=="Female"]["total_bill"].mean()
```

```python
#Visualization of invoices paid by incoming people according to days
```

```python
sns.barplot(x='day', y='total_bill',hue = "size", data=df)
```

```python
#visualize a histogram and density estimate of a data distribution
```

```python
sns.distplot(df["total_bill"])
plt.title('histogram and density plot')
```

```python
#The curve represents the density while the columns represent the histogram
```

```python
#Let's just consider the density plot
```

```python
sns.distplot(df.total_bill, hist = False)
```

```python
#Let's just consider the histogram chart
```

```python
sns.distplot(df.total_bill, kde = False)
```

```python
#The boxplot shows the median, quartiles (Q1 and Q3), and potential outliers.
```

```python
sns.boxplot(x = "day", y = "total_bill", hue = "smoker", data = df)
```

```python

sns.scatterplot(x = "total_bill", y = "tip", hue = "time", data = df)
```

```python
#Lmplot draws a regression line on the ScatterPlot plot, and this line can help us better understand 
#the relationship between two variables.
```

```python
sns.lmplot(x = "total_bill", y = "tip", hue = "smoker", data = df)
```

```python
#Pairplot is used to visualize the relationships between a data selected as reference and all numerical
#variables in the data set.
```

```python
sns.pairplot(df, hue = "total_bill")
```
